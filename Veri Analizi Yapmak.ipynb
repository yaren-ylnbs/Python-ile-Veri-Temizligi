```python
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean') #amaç NaN verileri tanımlamak yaş kolonunun ortalama verileri ile tamamla
```

```python
Yas #boy,kilo,yas
```

```python
imputer = imputer.fit(Yas[:,1:4])    #fit fonk eğitmek için kullanılır. bu değerleri öğren (strategy: mean -- ortalamasını öğren)
```

```python
Yas[:,1:4] = imputer.transform(Yas[:,1:4]) #öğrendikten sonra nan değerleri dönüştürmesini istiyoruz. bunun için de transform fonk kullanıyoruz.
```
